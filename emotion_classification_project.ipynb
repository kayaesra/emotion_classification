```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from huggingface_hub import login
import torch
device = torch.device("cuda")


tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-1.5B")
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen2.5-1.5B", device_map="auto")
valid_emotions = {"sadness", "joy", "love", "anger", "fear", "surprise"}
```

```python
label_map = {
    0: "sadness", 1: "joy", 2: "love",
    3: "anger", 4: "fear", 5: "surprise"
}

sample_y_labels = [label_map[label] for label in y_val]
```

```python

def classify_batch_with_prompt(text_batch):
    valid_emotions = {"sadness", "joy", "love", "anger", "fear", "surprise"}

    prompts = [
        f"""You are a helpful assistant. Classify the following sentence into one of the following emotions: sadness, joy, love, anger, fear, surprise.

        Sentence: "{text}"
        Emotion:"""
        for text in text_batch
    ]

    inputs = tokenizer(
        prompts,
        return_tensors="pt",
        padding="longest",
        truncation=True,
        max_length=128
    ).to(device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=10,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.pad_token_id
    )

    results = []
    for i in range(outputs.shape[0]):
        generated_tokens = outputs[i, inputs.input_ids.shape[1]:]
        result_text = tokenizer.decode(generated_tokens, skip_special_tokens=True)
        emotion_part = result_text.strip().lower()
        emotion = "unknown"
        tokens = emotion_part.split()
        clean_tokens = [token.strip(":,!.?'\"") for token in tokens]
        for token in clean_tokens:
            if token in valid_emotions:
                emotion = token
                break



        """if ":" in emotion_part:
            try: emotion = emotion_part.split(":")[-1].strip().split()[0].lower()
            except IndexError: pass
        elif emotion_part:
            try: emotion = emotion_part.split()[0].lower()
            except IndexError: pass"""


        results.append(emotion)
    return results


label_map = {
    0: "sadness", 1: "joy", 2: "love",
    3: "anger", 4: "fear", 5: "surprise"
}

sample_y_labels = [label_map[label] for label in y_val]



batch_size = 32
qwen_preds_val = []

print(f"\nToplam {len(x_val)} rnek ilenecek. Batch size: {batch_size}")
for i in range(0, len(x_val), batch_size):
    batch_texts = x_val[i:i+batch_size]
    current_batch_num = i // batch_size + 1
    total_batches = (len(x_val) + batch_size - 1) // batch_size
    print(f"leniyor batch {current_batch_num}/{total_batches} ({len(batch_texts)} rnek)...")


    batch_predictions = classify_batch_with_prompt(batch_texts)
    qwen_preds_val.extend(batch_predictions)

print("\nlem tamamland.")
```

```python
valid_qwen_preds_val = [p for p in qwen_preds_val if p != "unknown"]
valid_sample_y_labels_val = [sample_y_labels[i] for i, p in enumerate(qwen_preds_val) if p != "unknown"]

if not valid_qwen_preds_val:
    print("Hi geerli tahmin retilemedi.")
else:
    report_labels = list(label_map.values())
    print(classification_report(valid_sample_y_labels_val, valid_qwen_preds_val, labels=report_labels, zero_division=0))
```

```python
import pandas as pd

validation_set_dataframe = pd.DataFrame({
    'text': x_val,
    'true_label': sample_y_labels,
    'predicted_label': qwen_preds_val
})

incorrect_df_val = validation_set_dataframe[validation_set_dataframe['true_label'] != validation_set_dataframe['predicted_label']]
```

```python
confusion_counts = incorrect_df_val.groupby(['true_label', 'predicted_label']).size().unstack(fill_value=0)

print("\nHatal Tahmin Karklk Saylar:")
print(confusion_counts)
```

This indicates that the model has great difficulty or hesitations in choosing one of the 6 emotions specified for the given texts. Perhaps the prompt is not directive enough or the texts are ambiguous for the model.

->Looking at the surprise row, we see 0 predictions in the surprise column among the misclassified examples. Combined with the 0 00 F1 score from the classification report, this confirms that none of the validation examples with the true label "surprise" were correctly predicted as "surprise".
->Instead, 67 "surprise" examples were predicted as "unknown".
->The remaining "surprise" misclassifications were scattered among other labels: joy (1), love (3), and sadness (7).

Let's try to improve prompt. We gonna try few-shot prompting technique

```python
valid_emotions = {"sadness", "joy", "love", "anger", "fear", "surprise"}

few_shot_prefix = """You are a helpful assistant for classifying text into one of the following emotions:

sadness, joy, love, anger, fear, surprise.

You MUST respond with ONLY one of these words: sadness, joy, love, anger, fear, surprise.
No explanation. No other words.

Examples:

Sentence: "I feel so happy right now!"
Emotion: joy

Sentence: "This is the worst day of my life."
Emotion: sadness

Sentence: "I am so angry I could scream."
Emotion: anger

Sentence: "That sudden noise scared me half to death!"
Emotion: fear

Sentence: "I am deeply in love with this person."
Emotion: love

Sentence: "Wow, that was totally unexpected!"
Emotion: surprise
"""

def classify_with_prompt_FS(text_batch):
    prompts = [
        f"""{few_shot_prefix}

Sentence: "{text}"
Emotion:"""
        for text in text_batch
    ]

    inputs = tokenizer(
        prompts,
        return_tensors="pt",
        padding="longest",
        truncation=True,
        max_length=512
    ).to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=5,
        do_sample=False,
        num_beams=1,
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=tokenizer.eos_token_id
    )

    results = []
    for i in range(outputs.shape[0]):
        generated_tokens = outputs[i, inputs.input_ids.shape[1]:]
        result_text = tokenizer.decode(generated_tokens, skip_special_tokens=True).strip().lower()

        tokens = result_text.split()
        clean_tokens = [token.strip(":,!.?'\"") for token in tokens]
        emotion = "unknown"
        for token in clean_tokens:
            if token in valid_emotions:
                emotion = token
                break

        results.append(emotion)
    return results

label_map = {
    0: "sadness", 1: "joy", 2: "love",
    3: "anger", 4: "fear", 5: "surprise"
}

sample_y_labels = [label_map[label] for label in y_val]

batch_size = 32
qwen_preds_val_fs = []

print(f"\nToplam {len(x_val)} rnek ilenecek. Batch size: {batch_size}")
for i in range(0, len(x_val), batch_size):
    batch_texts = x_val[i:i + batch_size]
    current_batch_num = i // batch_size + 1
    total_batches = (len(x_val) + batch_size - 1) // batch_size
    print(f"leniyor batch {current_batch_num}/{total_batches} ({len(batch_texts)} rnek)...")

    batch_predictions = classify_with_prompt_FS(batch_texts)
    qwen_preds_val_fs.extend(batch_predictions)

print("\nlem tamamland.")
```

```python
valid_qwen_preds_val = [p for p in qwen_preds_val_fs if p != "unknown"]
valid_sample_y_labels_val = [sample_y_labels[i] for i, p in enumerate(qwen_preds_val_fs) if p != "unknown"]

if not valid_qwen_preds_val:
    print("Hi geerli tahmin retilemedi.")
else:
    report_labels = list(label_map.values())
    print("\n=== Classification Report For Few SHOT Prompting ===\n")
    print(classification_report(valid_sample_y_labels_val, valid_qwen_preds_val, labels=report_labels, zero_division=0))
```

```python
from tqdm import tqdm
from sklearn.metrics import classification_report

batch_size = 32
qwen_preds_Test = []

print(f"\nToplam {len(x_test)} rnek ilenecek. Batch size: {batch_size}")

for i in tqdm(range(0, len(x_test), batch_size), desc="leniyor (No FS)"):
    batch_texts = x_test[i:i + batch_size]
    batch_predictions = classify_batch_with_prompt(batch_texts)
    qwen_preds_Test.extend(batch_predictions)
```

```python

valid_qwen_preds_Test = [p for p in qwen_preds_Test if p != "unknown"]
valid_y_test = [label_map[y_test[i]] for i, p in enumerate(qwen_preds_Test) if p != "unknown"]

print("\n Classification Report (No Few-Shot):")
report_labels = list(label_map.values())
print(classification_report(valid_y_test, valid_qwen_preds_Test, labels=report_labels, zero_division=0))
```

```python
from tqdm import tqdm
from sklearn.metrics import classification_report

batch_size = 32
qwen_preds_FS_Test = []

print(f"\nToplam {len(x_test)} rnek ilenecek. Batch size: {batch_size}")

for i in tqdm(range(0, len(x_test), batch_size), desc="leniyor (Few-shot)"):
    batch_texts = x_test[i:i + batch_size]
    batch_predictions = classify_with_prompt_FS(batch_texts)
    qwen_preds_FS_Test.extend(batch_predictions)
```

```python
valid_qwen_preds_FS_Test = [p for p in qwen_preds_FS_Test if p != "unknown"]
valid_y_test_FS = [label_map[y_test[i]] for i, p in enumerate(qwen_preds_FS_Test) if p != "unknown"]

print("\nClassification Report (Few-Shot):")
report_labels = list(label_map.values())
print(classification_report(valid_y_test_FS, valid_qwen_preds_FS_Test, labels=report_labels, zero_division=0))
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np
from tqdm import tqdm

#for gpu
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

tokenizer2 = AutoTokenizer.from_pretrained("google-bert/bert-base-uncased")
model_TB = AutoModel.from_pretrained("google-bert/bert-base-uncased").to(device)
model_TB.eval()


def get_embedding(text):
    inputs = tokenizer2(text, return_tensors="pt", truncation=True, padding="max_length", max_length=64).to(device)
    with torch.no_grad():
        outputs = model_TB(**inputs)
    return outputs.last_hidden_state[:, 0, :].squeeze().cpu().numpy()

X_train = np.array([get_embedding(text) for text in tqdm(df_train["text"])])
y_train = df_train["label"].values

X_val = np.array([get_embedding(text) for text in tqdm(df_val["text"])])
y_val = df_val["label"].values

X_test = np.array([get_embedding(text) for text in tqdm(df_test["text"])])
y_test = df_test["label"].values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

#Klasik model
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)

#Tahmin ve metrikler
y_pred_val_emb = clf.predict(X_val)
print(" [Embedding + LogisticRegression] Validation Set Sonular:")
print(classification_report(y_val, y_pred_val_emb, target_names=list(label_map.values())))


y_pred_test_emb = clf.predict(X_test)
print("\n [Embedding + LogisticRegression] Test Set Sonular:")
print(classification_report(y_test, y_pred_test_emb, target_names=list(label_map.values())))
```

```python
train_data = Dataset.from_pandas(df_train)
val_data = Dataset.from_pandas(df_val)
test_data = Dataset.from_pandas(df_test)
```

```python
def tokenize_function(examples):
    return tokenizer2(examples['text'], padding="max_length", truncation=True)

train_data = train_data.map(tokenize_function, batched=True)
val_data = val_data.map(tokenize_function, batched=True)
test_data = test_data.map(tokenize_function, batched=True)

train_loader = DataLoader(train_data, batch_size=16)
val_loader = DataLoader(val_data, batch_size=16)
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"

from transformers import BertForSequenceClassification, TrainingArguments, Trainer
#(etiket says: 6)
model_cls = BertForSequenceClassification.from_pretrained("google-bert/bert-base-uncased", num_labels=6).to(device)

training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=64,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
    report_to=None
)

trainer = Trainer(
    model=model_cls,
    args=training_args,
    train_dataset=train_data,
    eval_dataset=val_data,
)

trainer.train()

```

```python
preds = trainer.predict(test_data)
y_pred = preds.predictions.argmax(axis=1)
y_true = test_data['label']

from sklearn.metrics import classification_report
print(classification_report(y_true, y_pred, target_names=list(label_map.values())))
```
